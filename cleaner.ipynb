```python
def change_to_province(code):
    if code >= 1000 and code < 1300:
        return "Brussel"
    elif code >= 1300 and code < 1500:
        return "Brabant Wallon","Wallonia"
    elif (code >= 1500 and code < 2000) or (code >= 3000 and code < 3500):
        return "Brabant Flamand","Flanders"
    elif code >= 2000 and code < 3000:
        return "Anvers","Flanders"
    elif code >= 3500 and code < 4000:
        return "Limbourg","Flanders"
    elif code >= 4000 and code < 5000:
        return "Liège","Wallonia"
    elif code >= 5000 and code < 6000:
        return "Namur","Wallonia"
    elif (code >= 6000 and code < 6600) or (code >= 7000 and code < 8000):
        return "Hainaut","Wallonia"
    elif code >= 6600 and code < 7000:
        return "Luxembourg","Wallonia"
    elif code >= 8000 and code < 9000:
        return "Flandre Occidental","Flanders"
    elif code >= 9000:
        return "Flandre Oriental","Flanders"
```

```python
data["Province"] = data.apply(lambda x: change_to_province(x["Locality"])[0], axis=1)
data["Region"] = data.apply(lambda x: change_to_province(x["Locality"])[1], axis=1)
```
